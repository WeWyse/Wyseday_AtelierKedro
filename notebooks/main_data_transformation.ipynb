```python
import pandas as pd

from typing import List

#Import kedro libraries
from kedro.config import TemplatedConfigLoader
from kedro.io import DataCatalog
```

```python
def sum_sales_by(
    """ method to sum sales values at group_key granularity """
      df: pd.DataFrame,
      group_key: List[str],
      aggregation_col: List[str]
):
    grouped_df = (df.copy()
                  .groupby(group_key, as_index=False)[aggregation_col]
                  .sum()
               )
    return grouped_df

def compute_price_by_liter(
    """ 
    method to calculate the price by litre of a drink brand 
    based on the sales values and volumes sold
    
    """
      df: pd.DataFrame,
      price_col_name: str,
      sales_col_name: str,
      volume_col_name: str
):
    df[price_col_name] = df[sales_col_name]/df[volume_col_name]
    return df
```
